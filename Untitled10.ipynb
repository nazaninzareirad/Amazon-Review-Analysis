1. Dataset Preparation

1.1 loading SECOM dataset

```python
import pandas as pd

# load the SECOM dataset (data part)
secom_data = pd.read_csv('/content/secom.data', header=None, sep=" ")

# display the first few rows of the dataset
secom_data.head()

num_features = secom_data.shape[1] # Get the number of columns from the DataFrame
column_names = [f'feature_{i}' for i in range(num_features)]

# set the columns in the dataframe
secom_data.columns = column_names

# display the columns
secom_data.columns

# load the labels (target) from the secom_labels.data file
secom_labels = pd.read_csv('/content/secom_labels.data', header=None)

# add the labels as the target column
secom_data['target'] = secom_labels

# display the first few rows with the target column
secom_data.head()
```

```python
# first, process the 'target' column to extract the numerical label
secom_data['target'] = secom_data['target'].apply(lambda x: int(str(x).split(' ')[0]))

# separate features (X) and target (y)
X = secom_data.drop(columns='target')
y = secom_data['target']

# replace missing values with the median of each feature column
X.fillna(X.median(), inplace=True)

#standardizing data
from sklearn.preprocessing import StandardScaler

# standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

2. classic methods

2.1 mutual information

```python
from sklearn.feature_selection import mutual_info_classif
import numpy as np

# calculate mutual information between features and target
mutual_info = mutual_info_classif(X_scaled, y)

# sort features based on mutual information
sorted_mi_features = np.argsort(mutual_info)[::-1]

# top 20 features based on mutual information
top_20_mi_features = sorted_mi_features[:20]
```

2.2 RFE

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE

# define the model for RFE
model = RandomForestClassifier(n_estimators=100, random_state=42)

rfe = RFE(model, n_features_to_select=20)
rfe.fit(X_scaled, y)

# get the ranking of features
rfe_ranking = rfe.ranking_

# top 20 features selected by RFE
top_20_rfe_features = np.argsort(rfe_ranking)[:20]
```
